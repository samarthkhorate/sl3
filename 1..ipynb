## Impute missing values

```python

categorical_cols = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status']

for col in categorical_cols:
    mode_value = df[col].mode()[0]  
    df[col].fillna(mode_value, inplace=True)  
```

```python

numerical_cols = ['Loan_Amount_Term', 'LoanAmount']

for col in numerical_cols:
    median_value = df[col].median()  
    df[col].fillna(median_value, inplace=True) 
```

```python
df['Credit_History'].fillna(df['Credit_History'].mean,inplace = True)
```

```python
df.isnull().sum()
```

```python
df
```

```python

df['Loan_ID'] = df['Loan_ID'].str[4:]
```

## Convert Categorical Variable to Numeric

```python
# converting object dtype (column) to string type
df['Loan_ID'] = df['Loan_ID'].astype(int)
```

```python
# replacing values
df["Gender"] = df["Gender"].replace(['Male', 'Female'],[0,1])
df.head(5)
```

```python
# Get dummies (One-Hot Encoding )
df = pd.get_dummies(df, columns=["Married", "Education", "Self_Employed", "Dependents"], drop_first=True)
df.head()
```

```python
Label = LabelEncoder()
df["Property_Area"]=Label.fit_transform(df["Property_Area"])
df
```
